```python
marks = [45, 78, 90, 33, 60]
p_count = 0
f_count = 0
for i in marks:
    if i >= 50:
        p_count += 1
    else:
        f_count += 1
print("Total number of students:", len(marks))
print("Total number of pass students:", p_count)
print("Total number of fail students:", f_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    name = name.strip() #removes space
    name = name.lower() #converts to lowercase
    cleaned_names.append(name) #add the cleaned name to new list
print("Cleaned Names", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    print("Messages:", msg)
    print("Length:", len(msg)) #len of msg
    if len(msg) > 10:
        print("Long message")
    print()
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for log in logs:
    if log == "ERROR":
        count += 1
print("The total error count:", count)
```
